```
import nltk
import pandas as pd
import string 
import re
from nltk.stem.snowball import SnowballStemmer
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```
df = pd.read_csv('ulasan.csv')
df.head()
```

###Lower Case

```
df.loc[:,"content"] = df.content.apply(lambda x : str.lower(x))
```

###Remove emoticon, and symbol

```
df.loc[:,"content"] = df.content.apply(lambda x : " ".join(re.findall('[\w]+',x)))
```

###Remove number

```
df.loc[:,"content"]
```

###Remove stopwords

```
factory = StopWordRemoverFactory()
stopword = factory.create_stop_word_remover()

df.loc[:,"content"] = df.content.apply(lambda x : stopword.remove(x))
```

###Stemming

```
factory = StemmerFactory()
stemmer = factory.create_stemmer()
df.loc[:,"content"] = df.content.apply(lambda x : stemmer.stem(x))
```

###Ubah data ke CSV baru

```
listdata = df['content'].values.astype('U')
listdata = [d for d in listdata]
```

```
df = pd.DataFrame(df)
df.to_csv('ulasan_fix.csv')
```
